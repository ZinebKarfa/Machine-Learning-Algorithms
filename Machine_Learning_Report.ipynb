# **Linear Regression**

Linear regression is a fundamental algorithm in the field of machine learning that plays a crucial role in modeling the relationship between a dependent variable and one or more independent variables. Its simplicity, interpretability, and wide applicability make it a popular choice in various domains, including finance, economics, social sciences, and many others. By understanding and implementing linear regression, we can gain valuable insights, make predictions, and uncover patterns within our data.

At its core, linear regression assumes a linear relationship between the independent variables (also known as features, predictors, or inputs) and the dependent variable (also known as the target, response, or output variable). The goal is to find the best-fit line that minimizes the overall distance between the predicted values and the actual values of the dependent variable. This process typically involves estimating the coefficients of the linear equation.

The coefficients derived from linear regression provide essential information about the relationship between the variables. The intercept coefficient represents the value of the dependent variable when all the independent variables are zero, providing a baseline reference point. The slope coefficients indicate the impact of each independent variable on the dependent variable, measuring the change in the dependent variable associated with a one-unit change in the corresponding independent variable, assuming the other variables remain constant.

Linear regression encompasses two main forms: simple linear regression and multiple linear regression. Simple linear regression involves a single independent variable and is suitable for analyzing the relationship between two variables. On the other hand, multiple linear regression extends the analysis to incorporate multiple independent variables, enabling a more comprehensive exploration of the relationships among various factors and the dependent variable.

Throughout this report, we will explore the fundamental concepts of linear regression and demonstrate its implementation using Python. By gaining a solid understanding of linear regression and its practical application, we can harness its power to uncover insights, predict outcomes, and make informed decisions in a wide range of real-world scenarios.

## **I. Simple Linear Regression**

Simple linear regression involves a single independent variable (X) and a dependent variable (Y). It assumes a linear relationship between X and Y, which can be represented by the equation: Y = b0 + b1*X. The coefficients b0 and b1 are estimated using the least squares method, minimizing the sum of squared residuals. Simple linear regression is particularly useful for analyzing and predicting relationships between two variables.

## **II. Multiple Linear Regression**

Multiple linear regression extends simple linear regression to include multiple independent variables. The equation becomes: Y = b0 + b1X1 + b2X2 + ... + bn*Xn. In this case, b0 represents the intercept, while b1 to bn are the coefficients for each independent variable. Multiple linear regression allows us to model more complex relationships by considering multiple factors simultaneously.

# **Principal Component Analysis (PCA)**

Principal Component Analysis (PCA) is a widely used dimensionality reduction technique in machine learning. It helps to transform a dataset with potentially high-dimensional features into a lower-dimensional space while preserving the most important information. By reducing the dimensionality, PCA simplifies data visualization, improves computational efficiency, and mitigates the curse of dimensionality. In this report, we will explore the concepts of PCA, its applications, and provide a Python code implementation to showcase its usage.

## **I. Understanding PCA**

PCA aims to find a new set of orthogonal variables, known as principal components, which capture the maximum variance in the data. It achieves this by projecting the original data onto these components. The first principal component captures the direction of maximum variance, while subsequent components capture the remaining variance in descending order. By selecting a subset of the principal components, we can represent the data in a lower-dimensional space.

## **II. Applications of PCA**

PCA has numerous applications in machine learning and data analysis:

*   **Dimensionality Reduction**: PCA is widely used to reduce the number of features while preserving the most important information in the data.
*   **Data Visualization**: By reducing the data to two or three dimensions, PCA enables effective visualization of high-dimensional data.
*   **Noise Reduction**: PCA can remove noise or irrelevant features by eliminating components with low variance.
*   **Preprocessing**: PCA is often used as a preprocessing step to improve the performance of other machine learning algorithms.
*   **Face Recognition**: PCA has been extensively used for face recognition tasks, extracting the most discriminative features from facial images.







## **III. Code Implementation**

Now, let's dive into the Python implementation of PCA using the scikit-learn library.

```python
import numpy as np
from sklearn.decomposition import PCA
```

# **Wavelet decomposition**

This unique property makes wavelet decomposition well-suited for handling non-stationary and transient signals, which are prevalent in many real-world applications.

In contrast, wavelet decomposition utilizes a set of wavelet functions that are localized both in time and frequency. This localization allows wavelet decomposition to capture localized features and transients within a signal more effectively.

In the context of machine learning, wavelet decomposition offers several advantages. It enables the extraction of meaningful features from signals, aiding in the improvement of classification and regression tasks. ## **I. Theory of Wavelet Decomposition**

***A. Overview of Fourier Transform and its limitations***

The Fourier Transform is a widely used mathematical tool for analyzing signals and decomposing them into their constituent frequencies. It provides a global frequency representation of the signal by decomposing it into a series of sinusoidal components. However, the Fourier Transform has limitations when it comes to analyzing signals with localized features or non-stationary characteristics. It does not capture time-varying behavior and lacks the ability to provide simultaneous time and frequency information.

***B. Introduction to wavelet functions and their properties***

Wavelet functions are the building blocks of wavelet decomposition. Unlike sinusoidal functions in Fourier Transform, wavelet functions are localized both in time and frequency domains. They possess properties such as compact support, oscillatory behavior, and zero average. These properties allow wavelet functions to capture localized features and transients present in signals, making them suitable for analyzing non-stationary and time-varying signals.

***C. Multiresolution analysis and the concept of scaling functions***

Multiresolution analysis is a key concept in wavelet decomposition. It involves representing a signal at different scales or resolutions. The decomposition is achieved through the use of scaling functions, which generate progressively smoother approximations of the original signal. Scaling functions capture the low-frequency components of the signal and provide a coarse representation at each level of the decomposition.

***D. Wavelet filters and filter banks***

Wavelet decomposition employs a set of wavelet filters that separate the input signal into different frequency bands. These filters operate in conjunction with downsampling, which reduces the number of samples while preserving the relevant information. The filters are designed to extract both low-frequency (approximation) and high-frequency (detail) components of the signal. Together, these filters and downsampling operations form a filter bank architecture for wavelet decomposition.

***E. Algorithmic steps of wavelet decomposition***

The algorithmic steps of wavelet decomposition involve iteratively applying wavelet filters and downsampling to the input signal. The process begins with the initial approximation and detail coefficients. The approximation coefficients capture the low-frequency components of the signal, while the detail coefficients capture the high-frequency components. These coefficients are further decomposed by applying the same filtering and downsampling operations iteratively to obtain the next level of approximation and detail coefficients.

***F. Comparison between wavelet decomposition and Fourier Transform***

Wavelet decomposition and Fourier Transform differ in their ability to analyze signals. Fourier Transform provides a global frequency representation but lacks time localization, making it less suitable for non-stationary signals. On the other hand, wavelet decomposition offers time and frequency localization, allowing for the analysis of localized features and time-varying behavior. Wavelet decomposition provides a more flexible and adaptive representation of signals, making it well-suited for various applications such as denoising, feature extraction, and time-frequency analysis.

In summary, wavelet decomposition overcomes the limitations of the Fourier Transform by utilizing wavelet functions that are localized in both time and frequency domains. Multiresolution analysis, wavelet filters, and downsampling form the core of wavelet decomposition. This technique provides simultaneous time and frequency information, making it suitable for analyzing non-stationary and time-varying signals. Comparing wavelet decomposition to Fourier Transform highlights the advantages of wavelet decomposition in capturing localized features and providing a more adaptive signal representation.

## **II. Wavelet Decomposition Algorithms**

***A. Continuous Wavelet Transform (CWT)***

The Continuous Wavelet Transform (CWT) is a wavelet decomposition algorithm that analyzes signals at all possible scales and positions. It involves convolving the input signal with a continuous family of wavelet functions, resulting in a continuous representation of the signal in both time and frequency domains. The CWT provides a time-scale representation, which allows for the detection of localized features at different scales. However, the CWT is computationally expensive due to the continuous nature of the transform.

***B. Discrete Wavelet Transform (DWT)***

The Discrete Wavelet Transform (DWT) is a computationally efficient version of wavelet decomposition that operates on discrete samples of a signal. It uses a discrete set of wavelet functions obtained through discretization of the continuous wavelet functions. The DWT decomposes a signal into different frequency subbands at multiple scales, providing a time-frequency representation of the signal. The DWT achieves this by applying a series of high-pass and low-pass filters to the signal, followed by downsampling. The DWT offers a multiresolution representation of the signal, enabling efficient analysis and reconstruction.

***C. Stationary Wavelet Transform (SWT)***

The Stationary Wavelet Transform (SWT) is an extension of the DWT that addresses one of its limitations, namely the shift-variance property. The SWT overcomes this limitation by applying the DWT to overlapping segments of the input signal. This results in a redundant representation of the signal, allowing for better time localization and preserving the shift-invariance property. The SWT is particularly useful for analyzing signals with non-stationary behavior and provides a high-resolution time-frequency representation.

***D. Wavelet Packet Transform (WPT)***

The Wavelet Packet Transform (WPT) is an extension of the DWT that offers a more flexible and detailed decomposition. It provides a complete binary tree structure, where each node in the tree represents a different frequency subband. The WPT allows for further decomposition of each subband into its own subbands, providing a comprehensive representation of the signal. The WPT offers increased flexibility in capturing both coarse and fine details of the signal, making it suitable for applications requiring a high level of signal analysis and feature extraction.

In summary, wavelet decomposition algorithms, including the Continuous Wavelet Transform (CWT), Discrete Wavelet Transform (DWT), Stationary Wavelet Transform (SWT), and Wavelet Packet Transform (WPT), provide different approaches to decomposing signals into time-frequency components. The CWT analyzes signals at all scales continuously, while the DWT operates on discrete samples and offers a computationally efficient multiresolution representation. The SWT addresses the shift-variance property of the DWT, enabling better time localization, and the WPT provides a more detailed and flexible decomposition by using a complete binary tree structure. The choice of algorithm depends on the specific requirements of the application and the desired level of time and frequency analysis.

##**III. Applications of Wavelet Decomposition in Machine Learning**

***A. Signal Denoising***

Wavelet decomposition is widely used for signal denoising tasks. By decomposing a noisy signal into different frequency subbands, wavelet decomposition allows for the separation of noise from the underlying signal components. Since wavelet functions are localized in both time and frequency domains, they effectively capture the local features of the signal, enabling the identification and removal of noise at specific scales. Wavelet denoising techniques, such as thresholding, exploit the sparsity of the wavelet coefficients to selectively remove noise while preserving the important signal components. This application finds relevance in various domains, including audio signal processing, image denoising, and biomedical signal analysis.

***B. Feature Extraction***

Wavelet decomposition is valuable for feature extraction in machine learning tasks. By decomposing a signal into different frequency subbands, wavelet decomposition provides a multi-resolution representation that captures both coarse and fine details of the signal. These decomposed subbands can serve as informative features for machine learning algorithms. For example, in image processing, wavelet coefficients at different scales and orientations can be utilized as features for object recognition, texture analysis, or image classification. Wavelet-based feature extraction can enhance the discriminative power of machine learning models by extracting relevant and localized information from the input signals.

***C. Image Compression***

Wavelet decomposition has proven to be effective in image compression algorithms. By decomposing an image into its frequency subbands, wavelet decomposition separates the image's high-frequency details from its low-frequency approximation. The high-frequency subbands, containing less perceptually significant information, can be quantized or discarded to achieve compression. The low-frequency approximation, which captures the global structure of the image, is retained at reduced resolution. This wavelet-based compression technique, such as the JPEG2000 standard, achieves efficient compression while preserving important image features and minimizing artifacts. Wavelet compression finds applications in various domains, including digital image storage, transmission, and multimedia systems.

***D. Time-Frequency Analysis***

Wavelet decomposition enables time-frequency analysis, which is essential for understanding the temporal evolution of frequency components in a signal. Unlike the Fourier Transform, which provides a global frequency representation, wavelet decomposition captures both time and frequency localization. This allows for the analysis of time-varying frequency components, transient events, and non-stationary signals. Time-frequency representations derived from wavelet decomposition, such as scalograms or spectrograms, provide insights into the time-varying spectral content of signals. This application is valuable in areas such as speech processing, audio analysis, vibration analysis, and physiological signal analysis, where understanding the time-varying frequency characteristics is crucial.

In summary, wavelet decomposition finds a wide range of applications in **machine learning**. It is used for **signal denoising** by separating noise from the underlying signal components. Wavelet decomposition enables **feature extraction** by capturing informative subbands for machine learning tasks. It plays a significant role in **image compression** algorithms by decomposing images into frequency subbands and achieving efficient compression. Additionally, wavelet decomposition facilitates **time-frequency analysis**, allowing for the characterization of time-varying frequency components in signals. These applications highlight the versatility and effectiveness of wavelet decomposition in enhancing machine learning tasks and signal processing applications.

```python
import yfinance as yf
import numpy as np
import pywt
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.stattools import adfuller
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import datetime
import matplotlib.pyplot as plt
```

### ***Load and preprocess the stock price data***

```python
# Download the stock price data from Yahoo Finance
symbol = 'META'  # META Stock
start_date = '2010-01-01'  # Define the start date
end_date = '2023-06-26'  # Define the end date
stock_data = yf.download(symbol, start=start_date, end=end_date)

# Extract the adjusted close prices as the time series
time_series = stock_data['Adj Close'].values

# Apply any necessary preprocessing steps (e.g., removing outliers, handling missing values)
# For simplicity, we'll assume no preprocessing is required in this example.

# Visualize the adjusted close prices
plt.figure(figsize=(10, 6))
plt.plot(time_series)
plt.title("Adjusted Close Price META")
plt.xlabel("Date")
plt.ylabel("Adjusted Close Price")
plt.grid(True)
plt.show()
```

### ***Wavelet decomposition***

```python
# Convert the data to list
data = time_series.tolist()

# Choose the wavelet function and decomposition level
wavelet = 'db4'
level = 5

# Perform wavelet decomposition on the time series
coefficients = pywt.wavedec(data, wavelet, level=level)

#Reconstruct the approximation coefficients and details
approximation = coefficients[0]
details = coefficients[1:]

#Visualize details
plt.figure(figsize=(10, 6))
for i, detail_level in enumerate(details):
    plt.subplot(len(details), 1, i+1)
    plt.plot(detail_level)
    plt.title(f"Detail Level {i+1}")
plt.tight_layout()
plt.show()
```

### ***AR model and Prediction for each detail level***

```python
# Prediction of future values for each detail level using AR model
for i, detail_level in enumerate(coefficients[1:]):
    model = AutoReg(detail_level, lags=10) # Determine the order of the AR model (lags)
    model_fit = model.fit()  # Fitting the model to the data

    # Make Predictions
    predictions = model_fit.predict(start=len(detail_level), end=len(detail_level) + 10)

    # Visualize Predictions
    plt.figure(figsize=(10, 6))
    plt.plot(detail_level, label="Historical Data")
    plt.plot(range(len(detail_level), len(detail_level) + len(predictions)), predictions, label="Predictions")
    plt.title(f"Predictions of the META Stock Price - Detail Level {i+1}")
    plt.xlabel("Time")
    plt.ylabel("META Price")
    plt.legend()
    plt.show()
```

### ***Stationarity Analysis : ADF test***

```python
# Stationarity Analysis for each detail level
for i, detail_level in enumerate(coefficients[1:]):
    # Augmented Dickey-Fuller Test
    result = adfuller(detail_level)
    # P-value
    p_value = result[1]
    # Read the result
    if p_value < 0.05:
        stationarity = "Stationnary"
    else:
        stationarity = "Non-stationnary"
    print(f"Detail Level {i+1} : {stationarity} (p-value: {p_value})")
```

### ***Linear Regression and Predictions***

```python
# Prediction of future values for each detail level
predictions_combined = []
for i, detail_level in enumerate(coefficients[1:]):
    # Autregressive Model (AR) with a lag of 10
    model = LinearRegression()
    model.fit(np.arange(len(detail_level)).reshape(-1, 1), detail_level)
    # Prediction of the AR model
    predictions = model.predict(np.arange(len(detail_level), len(detail_level) + 10).reshape(-1, 1))
    # Combine the predictions
    predictions_combined.append(predictions)

# Transform the combined predictions to a 2D array
predictions_combined = np.array(predictions_combined).T

# Repeat the predictions to adapt to the length of the original time series
repeated_predictions = np.repeat(predictions_combined, len(data) // len(predictions_combined) + 1, axis=0)
repeated_predictions = repeated_predictions[:len(data)]

# Linear Regression to estimate the original time series
regressor = LinearRegression()
regressor.fit(repeated_predictions, data)

# Final prediction of the original time series
original_timeseries_predictions = regressor.predict(repeated_predictions)
```
